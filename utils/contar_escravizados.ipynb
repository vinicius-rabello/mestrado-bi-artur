```python
import pandas as pd
from tqdm import tqdm
```

```python
df = pd.read_excel('../database/tables/Censo31_powerbi.xlsx')
df['Escravizados'] = None
```

```python
dono_atual, n_escravizados = 0, None
for i in tqdm(range(df.shape[0])):
    relacao, condicao, distrito = df.loc[i][['Relação', 'Condição', 'Distrito']]
    # checar se o indice atual é de um dono
    if relacao == 100:
        if dono_atual is not None:
            df.loc[dono_atual, 'Escravizados'] = n_escravizados
        # se distrito for nulo, dono não conta
        if isinstance(distrito, str):
            dono_atual = i
        else:
            dono_atual = None
        n_escravizados = 0
    elif condicao == 'escravo':
        n_escravizados +=1
```

```python
df.to_excel('../database/tables/Censo31_escravizados.xlsx', index=False)
```
