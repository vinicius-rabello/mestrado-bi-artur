```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('../database/tables/Censo31_escravizados.xlsx')
```

```python
temp = df[~df['Escravizados'].isna()]['Escravizados'].reset_index(drop=True)
total_chefes = temp.shape[0]
total_escravos = temp.sum()
print(total_chefes, total_escravos)
```

```python
tabela = pd.DataFrame(data=['0', '01', '02', '03-05', '06-10', '11-15', '16-20', '21-30', '+30', 'N'], columns=['Tamanho do Plantel'])
tabela
```

```python
count_chefes = []
count_escravos = []
for tamanho in tabela['Tamanho do Plantel']:
    match tamanho:
        case '0':
            temp = df[df['Escravizados'] == 0]
            n_chefes = temp.shape[0]
            n_escravos = temp['Escravizados'].sum()
            count_chefes.append(n_chefes)
            count_escravos.append(n_escravos)
        case '01':
            temp = df[df['Escravizados'] == 1]
            n_chefes = temp.shape[0]
            n_escravos = temp['Escravizados'].sum()
            count_chefes.append(n_chefes)
            count_escravos.append(n_escravos)
        case '02':
            temp = df[df['Escravizados'] == 2]
            n_chefes = temp.shape[0]
            n_escravos = temp['Escravizados'].sum()
            count_chefes.append(n_chefes)
            count_escravos.append(n_escravos)
        case '03-05':
            temp = df[(df['Escravizados'] >= 3) & (df['Escravizados'] <= 5)]
            n_chefes = temp.shape[0]
            n_escravos = temp['Escravizados'].sum()
            count_chefes.append(n_chefes)
            count_escravos.append(n_escravos)
        case '06-10':
            temp = df[(df['Escravizados'] >= 6) & (df['Escravizados'] <= 10)]
            n_chefes = temp.shape[0]
            n_escravos = temp['Escravizados'].sum()
            count_chefes.append(n_chefes)
            count_escravos.append(n_escravos)
        case '11-15':
            temp = df[(df['Escravizados'] >= 11) & (df['Escravizados'] <= 15)]
            n_chefes = temp.shape[0]
            n_escravos = temp['Escravizados'].sum()
            count_chefes.append(n_chefes)
            count_escravos.append(n_escravos)
        case '16-20':
            temp = df[(df['Escravizados'] >= 16) & (df['Escravizados'] <= 20)]
            n_chefes = temp.shape[0]
            n_escravos = temp['Escravizados'].sum()
            count_chefes.append(n_chefes)
            count_escravos.append(n_escravos)
        case '21-30':
            temp = df[(df['Escravizados'] >= 21) & (df['Escravizados'] <= 30)]
            n_chefes = temp.shape[0]
            n_escravos = temp['Escravizados'].sum()
            count_chefes.append(n_chefes)
            count_escravos.append(n_escravos)
        case '+30':
            temp = df[df['Escravizados'] > 30]
            n_chefes = temp.shape[0]
            n_escravos = temp['Escravizados'].sum()
            count_chefes.append(n_chefes)
            count_escravos.append(n_escravos)

count_chefes = [round(100*count/total_chefes, 2) for count in count_chefes]
count_escravos = [round(100*count/total_escravos, 2) for count in count_escravos]
count_chefes.append(total_chefes)
count_escravos.append(total_escravos)
print(count_chefes)
print(count_escravos)
```

```python
tabela.insert(loc=1, column='Chefes de Domicílio (%)', value = count_chefes)
tabela.insert(loc=2, column='Escravos Possuídos (%)', value = count_escravos)
```
